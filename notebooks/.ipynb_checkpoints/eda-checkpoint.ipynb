## Stepwise Selection

```python

# Forward stepwise selection
def forward_stepwise(X, y, max_features=8):
    selected = []
    remaining = list(X.columns)
    
    for _ in range(max_features):
        best_score = -np.inf
        best_feature = None
        
        for feature in remaining:
            temp_features = selected + [feature]
            temp_model = sm.OLS(y, X[temp_features]).fit()
            
            if temp_model.rsquared > best_score:
                best_score = temp_model.rsquared
                best_feature = feature
        
        if best_feature is None:
            break
            
        selected.append(best_feature)
        remaining.remove(best_feature)
        print(f"Added {best_feature} (R²={best_score:.3f})")
    
    return selected

target = "Exam_Score"
X = df.drop(target, axis=1)
y = df[target]

X = pd.get_dummies(X, drop_first=True)
X = X.astype(float)  # This fixes the bool problem

print("All numeric now:", (X.dtypes == 'float64').all())

X = sm.add_constant(X)
best_features = forward_stepwise(X, y)

final_model = sm.OLS(y, X[best_features + ['const']]).fit()
print(final_model.summary())
```

## Top Features

```python
importance_df = pd.DataFrame({
    'Feature': best_features,
    'Coef': final_model.params[best_features]
}).sort_values('Coef', ascending=False).head(8)

plt.figure(figsize=(12, 6))
sns.barplot(data=importance_df, x='Coef', y='Feature')
plt.title("Top Features Predicting Exam Score\n(R²=0.611)", fontsize=16, fontweight='bold')
plt.xlabel("Coefficient")
plt.tight_layout()
plt.savefig("feature_importance.png", dpi=300, bbox_inches='tight')
plt.show()

# 2. Attendance vs score (your strongest finding)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Attendance', y='Exam_Score', alpha=0.6)
sns.regplot(data=df, x='Attendance', y='Exam_Score', scatter=False, color='red', line_kws={'linewidth':3})
plt.title("Attendance is the #1 Predictor\n(+0.199 points per % increase)", fontsize=14, fontweight='bold')
plt.savefig("attendance_effect.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Outliers

```python
# Boxplots for top features
top_features = ['Attendance', 'Previous_Scores', 'Hours_Studied']
for col in top_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, y=col)
    plt.title(f"Outliers in {col}")
    plt.tight_layout()
    plt.show()

# IQR outliers count
for col in top_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1-1.5*IQR) | (df[col] > Q3+1.5*IQR)].shape[0]
    print(f"{col}: {outliers} outliers ({outliers/len(df)*100:.1f}%)")
```

## VIF 

```python
# Get numeric stepwise features that exist in original df
top_numeric_features = [f for f in best_features[:6] if f in df.columns]
print("Using these numeric features:", top_numeric_features)

# Only use features that exist
X_vif = sm.add_constant(df[top_numeric_features].astype(float))

# VIF calculation
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print("✅ VIF scores (<5 = good):")
print(vif_data.sort_values('VIF', ascending=False))

plt.figure(figsize=(10, 6))
sns.barplot(data=vif_data, x='VIF', y='feature')
plt.axvline(x=5, color='red', linestyle='--', label='VIF=5')
plt.title("Multicollinearity Check - All Good!")
plt.legend()
plt.tight_layout()
plt.savefig("vif_final.png", dpi=300)
plt.show()
```
